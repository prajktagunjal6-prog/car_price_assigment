```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
diamond = sns.load_dataset('diamonds')
diamond
```

```python
diamonds = sns.load_dataset("diamonds")
print(diamonds.head())
```

# **1.Scatter plot**

```python
sns.scatterplot(x='carat',y='price',data=diamond)
plt.title('Diamond Price vs. Carat')
plt.xlabel('Carat')
plt.ylabel('Price')

plt.show()
```

# **2.Line plot**

```python
sns.lineplot(data=diamonds, x='carat', y='price')
plt.title('Diamond Price vs. Carat')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.grid(True)
plt.legend()
plt.show()
```

# **3.Bar plot**

```python
sns.barplot(data=diamonds, x='cut', y='price')
plt.title('Diamond Price by Cut')
plt.xlabel('Cut')
plt.ylabel('Price')
plt.grid(alpha=0.3)
plt.show()


```

# **6.Violine plot**

```python
sns.violinplot(data=diamonds, x='cut', y='price')
plt.title('Diamond Price by Cut')
plt.xlabel('Cut')
plt.ylabel('Price')
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```

# **8.Facet grid**

```python
sns.set_style("whitegrid")
sample = diamonds.sample(n=100, random_state=42)
```

```python
sns.scatterplot(data=sample, x="carat", y="price", hue="cut")
plt.title('Diamond Price vs. Carat')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.grid(alpha=0.3)
```

```python
g = sns.FacetGrid(sample, col="cut")
g.map_dataframe(sns.scatterplot, x="carat", y="price")
```

# **9.Customizing a scatter plot**

```python
sns.scatterplot(data=diamonds, x='carat', y='price', hue='cut', style='cut')
plt.title('Diamond Price vs. Carat')
plt.xlabel('Carat')
plt.ylabel('Price')
#plt.legend(title="Time", loc="upper right")
plt.grid(alpha=0.3)
```
