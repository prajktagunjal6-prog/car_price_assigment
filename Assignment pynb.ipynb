```python
import pandas as pd
import numpy as np
```

```python
#load this file
df = pd.read_csv("/content/car_price_dataset.csv")
df
```

```python
# What is the average price of all cars in the dataset?
print(f"The average price of all cars in the dataset is ${round(df['Price'].mean(),2)}")
```

```python
#2. Which car brand has the highest average price?

Brand =df[df["Price"] == df["Price"].max()]["Brand"][1100]

print(f"The brand with the highest price is {Brand}")
```

```python
#3.What is the most common fuel type in the dataset?
most_common_fuel = df["Fuel_Type"].mode()[0]
print(f"The most common fuel type {most_common_fuel}")
```

```python
#4. How many cars have a mileage greater than 200,000?
count = (df['Mileage'] > 200000).sum()
print("Number of cars with mileage greater than 200,000:", count)
```

```python
#5. What is the distribution of cars by transmission type?

transmission_distribution = df["Transmission"].value_counts()
print(transmission_distribution)
```

```python
#6. Which car model has the highest price?
highest_price_car = df.loc[df["Price"].idxmax()]

print("Car Model:", highest_price_car["Model"])
print("Brand:", highest_price_car["Brand"])
print("Price:", highest_price_car["Price"])
```

```python
#7. What is the average price of cars by fuel type?
average_price = df.groupby("Fuel_Type")["Price"].mean()

print(average_price)
```

```python
#8. How many cars are from the year 2020 or later?
count = df[df["Year"] >= 2020].shape[0]

print("Number of cars from 2020 or later:", count)
```

```python
#9. What is the average engine size for each brand?
average_engine_size = df.groupby("Brand")["Engine_Size"].mean()

print(average_engine_size)
```

```python
#10. Which car has the lowest mileage?
lowest_mileage_car = df.loc[df["Mileage"].idxmin()]


print(lowest_mileage_car)
```

```python
#11. What is the correlation between mileage and price?
correlation = df["Mileage"].corr(df["Price"])

print("Correlation between Mileage and Price:", correlation)

```

```python
#12. How many cars have more than 2 owners?

count = df[df["Owner_Count"] > 2].shape[0]

print("Number of cars with more than 2 owners:", count)

```

```python
#13. What is the average price of cars by year?
average_price_by_year = df.groupby("Year")["Price"].mean()

print(average_price_by_year)
```

```python
#14. Which car has the highest number of doors?
highest_doors_car = df.loc[df["Doors"].idxmax()]

print(highest_doors_car)
```

```python
#15. What is the average mileage for each fuel type?
average_mileage_by_fuel = df.groupby("Fuel_Type")["Mileage"].mean()

print(average_mileage_by_fuel)
```

```python
#16. How many cars are from the brand 'Toyota'?
toyota_count = (df["Brand"] == "Toyota").sum()

print("Number of Toyota cars:", toyota_count)
```

```python
#17. What is the average price of cars with automatic transmission?
average_price = df[df["Transmission"] == "Automatic"]["Price"].mean()

print("Average price of cars with automatic transmission:", average_price)
```

```python
#18. Which car has the highest owner count?
highest_owner_count_car = df.loc[df["Owner_Count"].idxmax()]

print(highest_owner_count_car)
```

```python
#19. What is the average engine size for cars with diesel fuel type?
average_engine_size = df[df["Fuel_Type"] == "Diesel"]["Engine_Size"].mean()

print("Average engine size for cars with diesel fuel type:", average_engine_size)
```

```python
#20. How many cars have a price greater than $10,000?


count = (df["Price"] > 10000).sum()

print("Number of cars with price greater than $10,000:", count)
```
